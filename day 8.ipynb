```python
file= open(filename, 'r', encoding='latin-1')
L = []
first_line=file.readline()
for x in file:
    x_list = x.split('\t')
    try:
        yt_likes = int(float(x_list[12]))
        album_name = x_list[1]
        L.append(  (yt_likes, album_name)  )

    except ValueError:
        pass 
 
file.close()
```

```python
#what is the most liked album on yt
#print(album_list_yt_likes)
i = 0
current_max = 0
cm_idx = 0
for idx1 in range(len(L)):
    L2 = L[idx1]
    for idx2 in range(len(L2)):
        if L2[0] > current_max:
            current_max = L2[0]
            cm_idx = idx1
    i += 1

print(current_max, cm_idx)

```
