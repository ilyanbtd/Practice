```python
# create a circle plot
def draw_line(u, v):
    (x1, y1) = u
    (x2, y2) = v
    plt.plot( [x1, x2], [y1, y2], 'k')
```

```python
draw_line([1, 3], [5, 3])
draw_line([1, 1], [5, 1])


draw_line([1, 1], [1, 3])
draw_line([5, 3], [5, 1])
```
